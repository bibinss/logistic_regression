```python
from pandas import DataFrame, read_csv
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
import random
from sklearn import datasets, linear_model
from sklearn.linear_model import LogisticRegression
```

# Linear Regression

```python
df = pd.read_excel('housing_price.xlsx')
```

**Plot input data points**

```python
plt.scatter(df['area'], df['price'], color='b')
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```

```python
area = df['area'].to_numpy()[:, np.newaxis]
price = df['price'].to_numpy()[:, np.newaxis]
```

**Use the Linear Regression from scikit learn library and train the model**

```python
regr = linear_model.LinearRegression()
regr.fit(area, price)
```

**Plot the line**

```python
price_y_pred = regr.predict(area)
plt.plot(df['area'], price_y_pred, color='r')
plt.scatter(df['area'], df['price'], color='b')
plt.xlabel("area")
plt.ylabel("price")
plt.show()
```

# **Logistic Regression**

**Plot the points - blue circles and green triangles**

```python
num_data_points = 200
# generate input points for both classes
blue_circle_points = np.random.normal(2, .5, (num_data_points, 2))
green_triangle_points = np.random.normal(4, .5, (num_data_points, 2))
plt.figure(figsize=(8, 6), dpi=80)
plt.scatter(blue_circle_points[:,0], blue_circle_points[:,1], marker='o', s=100, color='b', label='Class:0')
plt.scatter(green_triangle_points[:,0], green_triangle_points[:,1], marker='^', s=100, color='g', label='Class:1')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

```python
label_blue_circle = np.zeros(num_data_points, dtype=int)
label_green_triangle = np.ones(num_data_points, dtype=int)
X = np.concatenate((blue_circle_points, green_triangle_points))
y = np.concatenate((label_blue_circle, label_green_triangle))
```

**Use LogisticRegression from scikit learn library and train the model**

```python
clf = LogisticRegression(random_state=0).fit(X, y)
```

**Plot the boundary line**

```python
b = clf.intercept_[0]
w1, w2 = clf.coef_.T
# Calculate the intercept and gradient of the decision boundary.
c = -b/w2
m = -w1/w2
xmin, xmax = 0, 6
ymin, ymax = 0, 6
xd = np.array([xmin, xmax])
yd = m*xd + c
plt.figure(figsize=(8, 6), dpi=80)
plt.plot(xd, yd, color='r')
plt.scatter(blue_circle_points[:,0], blue_circle_points[:,1], marker='o', s=100, color='b')
plt.scatter(green_triangle_points[:,0], green_triangle_points[:,1], marker='^', s=100, color='g')
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```

**Logistic function / Sigmoid function**
